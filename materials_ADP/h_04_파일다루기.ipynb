## <a id='toc1_1_'></a>[파일 열기](#toc0_)

```python
# 쓰기 전용으로 파일 열기 (write)
# 파일이 없으면 새로 만든다
f = open("test.csv", "w")
f.write("a,b,c,d,e \n")
f.write("1,2,3,4,5 \n")
f.close()
```

```python
# 파일 내용 보기
with open('test.csv') as f:
    contents = f.read()
    print(contents)
```

### <a id='toc1_1_1_'></a>[with 사용](#toc0_)
- 파일을 열 때 with를 사용하면, 해당 블록을 나갈 때 열린 파일을 닫는다
- 즉, close 호출을 잊어버려도 안전하게 파일을 닫는다

```python
# 운영체제 명령으로 보기
# !cat test.csv  # 리눅스나 맥에서
# !type test.csv # 윈도우에서
```

### <a id='toc1_1_2_'></a>[파일 오픈 옵션](#toc0_)
'r'	읽기 (디폴트)  
'w'	쓰기 (파일이 없었으면 새로 생성하며 이미 있었으면 덮어쓴다)  
'a'	맨 뒤에 추가 (appending), 파일이 없었으면 새로 생성  
't'	텍스트 모드로 오픈 (디폴트)  
'b'	바이너리 모드로 오픈  
'+'	읽기 쓰기 모두 가능  

```python
# 기존 파일에 추가하는 옵션으로 파일 열기 (append)
with open("test.csv", "a") as f:
    f.write("6,7,8,9,10 \n")
    f.write("11,12,13,14,15 \n")
    f.write("16,17,18,19,20 \n")
```

```python
with open('test.csv') as f: 
    print(f.read())
```

### <a id='toc2_2_4_'></a>[walk() 사용하기](#toc0_)
- os.walk()를 사용하면 경로, 폴더, 파일명을 한번에 찾아준다

```python
# 모든 하위의 하위 폴더를 검색하여 확장자가 .ipynb인 파일을 찾는다
for (path, dir, files) in os.walk(cur_dir):
    print("\n")
    print("경로: ", path)
    print("폴더: ", dir)
    print("모든 파일:")
    for file_name in files:
        extention = os.path.splitext(file_name)[-1]
        if extention == ".ipynb":
            print(file_name)
```

# <a id='toc3_'></a>[파일 생성](#toc0_)
- 임의의 개인정보 생성

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
# 참고도서
"""
Author : Byunghyun Ban
GitHub : https://github.com/needleworm
Book : 6개월 치 업무를 하루 만에 끝내는 업무 자동화
Last Modification : 2020.02.12.
"""
import time
import random
import os


# 생성할 개인정보 파일수
NUM_SAMPLES = 1000

# 이메일 생성에 사용할 샘플 글자 범위
alphabet_samples = "abcdefghizklmnopqrstuvwxyz1234567890"

# 임의의 길이의 영어 글자 랜덤 선택
def random_string(length):
    result = ""
    for i in range(length):
        result += random.choice(alphabet_samples)
    return result


# 이름 생성에 사용할 샘플 글자들
first_name_samples = "김이박최정강조윤장임"
middle_name_samples = "민서예지도하주윤채현지"
last_name_samples = "준윤우원호후서연아은진"


# 무작위로 사람 이름을 생성하는 함수
def random_name():
    result = ""
    result += random.choice(first_name_samples)
    result += random.choice(middle_name_samples)
    result += random.choice(last_name_samples)
    return result
```

```python
# 현재의 위치를 확인
os.getcwd()
```

```python
# 결과물을 저장할 폴더가 이미 존재하면 폴더를 지우고 새로 폴더를 만든다
# 기존 내용을 모두 지우기 위함
import shutil
if os.path.exists("./personal_info"):
    shutil.rmtree('./personal_info')
os.mkdir("./personal_info")    
```

```python
from random import randint
randint(25,99)
```

```python
# 개인정보 파일 자동 생성
for i in range(NUM_SAMPLES):
    # 무작위로 사람 이름 생성
    name = random_name()
    # 저장할 파일 이름 정의
    filename = "personal_info/" + str(i) + "_" + name + ".txt"
    # 빈 파일 생성
    outfile = open(filename, 'w')
    # 결과물 파일에 이름을 기재
    outfile.write("name : " + name + "\n")
    # 결과물 파일에 무작위로 생성된 나이를 기재
    outfile.write("age : " + str(str(randint(25,99)) + "\n"))
    # 결과물 파일에 무작위로 생성된 이메일을 기재
    outfile.write("e-mail : " + random_string(8) + "@bhban.com\n")
    # 결과물 파일에 무작위로 생성된 부서명을 기재
    outfile.write("division : " + random_string(3) + "\n")
    # 결과물 파일에 무작위로 생성된 핸드폰 번호를 기재
    outfile.write("telephone : 010-" + str(randint(1000,9999)) + "-" + str(randint(1000,9999)) + '\n')
    # 결과물 파일에 무작위로 선정된 성별을 기재
    outfile.write("sex : " + random.choice(["male", "female"]))
    outfile.close()

print("생성된 파일수", len(os.listdir("./personal_info/")))
```

# <a id='toc4_'></a>[파일 합치기](#toc0_)

```python
directory = "personal_info"

# 결과를 저장할 파일명 (txt 또는 csv 선택)
# outfile_name = "merged_ID.txt"
outfile_name = "merged_ID.csv"
out_file = open(outfile_name, 'w')

for filename in file_list:
    # 텍스트 파일이 아니면 건너뛴다
    if ".txt" not in filename:
        continue
    file = open(directory + "/" + filename)
    content = file.read()
    out_file.write(content + "\n\n")
    file.close()

out_file.close()
```

```python
# 결과 모두 출력
with open(outfile_name) as f: 
    print(f.read())
```

```python
# 일부 라인만 얻기, readline() 사용
number_of_lines = 10

out = []
with open(outfile_name) as f: 
    for i in range(number_of_lines):
        line = f.readline()
        out.append(line)
        print(line)
out
```

```python
# 일부 라인만 얻기, readlines() 사용

with open(outfile_name) as f: 
    out = f.readlines()
out[:number_of_lines]
```

```python
# 일부 라인만 읽기, next() 사용
with open(outfile_name) as f: 
    out = [next(f) for i in range(number_of_lines)]
out
```

## <a id='toc5_1_'></a>[ 헤더 만들기](#toc0_)
- 테이블 구조로 만들기
- 컬럼명으로 name, age, e-mail, ... 등을 지정하는 방법

```python
outfile_name = "merged_ID.csv"
out_file = open(outfile_name, 'w', encoding="euc_kr")
headers = []
outfile_has_header = False

for filename in file_list:
    if ".txt" not in filename:
        continue
    file = open(directory + "/" + filename)
    contents = []
    for line in file:
        # 엑셀 파일의 양식과 내용물을 분리합니다.
        if ":" in line:
            splits = line.split(":")
            contents.append(splits[-1].strip())

            # 헤더를 정리합니다. 최초 1회만 실행됩니다.
            if len(contents) > len(headers):
                headers.append(splits[0].strip())

    # 헤더를 파일에 입력합니다. 최초 1회만 실행됩니다.
    if not outfile_has_header:
        header = ", ".join(headers)
        out_file.write(header)
        outfile_has_header = True

    # 결과물 파일에 내용물을 입력합니다.
    new_line = ", ".join(contents)
    out_file.write("\n" + new_line)

    # 읽어온 파일을 종료합니다.
    file.close()

# 결과물 파일을 종료합니다.
out_file.close()
```
